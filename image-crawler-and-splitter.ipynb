```python
import splitfolders
from icrawler.builtin import GoogleImageCrawler
```

```python
food_classes = ['roti-kukus-mekar', 'lumpia', 'kue-lapis']
```

```python
for food in food_classes:
    crawler = GoogleImageCrawler(storage = {'root_dir': f'data/{food}'})
    crawler.crawl(keyword = f'{food}', min_size=(100, 100), max_num = 100)
```

```python
splitfolders.ratio('data', output="output", seed=61, ratio=(.8, 0.1,0.1)) 
```
